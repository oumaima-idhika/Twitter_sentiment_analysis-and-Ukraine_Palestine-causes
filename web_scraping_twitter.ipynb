```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
```

```python
def getData(query):
    tweets = []
    limit = 5000
    for tweet in sntwitter.TwitterSearchScraper(query).get_items():
        tweets
        if len(tweets) == limit:
            break
        else:
            tweets.append([tweet.id,tweet.date,tweet.user.username,tweet.replyCount,tweet.retweetCount,tweet.likeCount,tweet.lang,tweet.content])
    return tweets
```

```python
df_ukraine = pd.DataFrame(getData('ukraine'), columns=['ids','date','user','replyCount','retweetCount','likeCount','lang','text'])
df_palestine=pd.DataFrame(getData('palestine'), columns=['ids','date','user','replyCount','retweetCount','likeCount','lang','text'])
```
